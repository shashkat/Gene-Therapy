```python
import pandas as pd
```

```python
data = pd.read_csv('fastq_barcode_matrix.csv')
data
```

```python
# make a dictionary of barcode names

# GTGCATGA		AAV92YF
# TGTTGGAG		Rh10
# GACATACG		ATX002
# GAGTAACG	    ATX001
# GAGAGCAA	    AAV9
# GGTGTTAG	    7m8
# ACGAACGA	    AAV2
# CGAAGCAA	    AAV8
# GTCCTGAA	    AAV6

barcode_names = {'GTGCATGA': 'AAV92YF', 'TGTTGGAG': 'Rh10', 'GACATACG': 'ATX002', 'GAGTAACG': 'ATX001', 'GAGAGCAA': 'AAV9', 'GGTGTTAG': '7m8', 'ACGAACGA': 'AAV2', 'CGAAGCAA': 'AAV8', 'GTCCTGAA': 'AAV6'}
```

```python
# create a barcode_names_array array
barcode_names_array = []
for barcode in data['barcodes']:
    barcode_names_array.append(barcode_names[barcode])
# add the barcode_names_array to the data and put it in the second column
data.insert(1, 'barcode_names', barcode_names_array)
data
```

```python
# below is the information about the initial numbers for each barcode's virus 
# notice the discrepancy between the order of the barcodes in the two replicates

# GTGCATGA	1029270		AAV92YF		GTGCATGA	868583		AAV92YF
# TGTTGGAG	365845		Rh10		TGTTGGAG	378761		Rh10
# GACATACG	293252		ATX002		GACATACG	274134		ATX002
# GAGTAACG	138779	    ATX001		GAGTAACG	154330	    ATX001
# GAGAGCAA	127729	    AAV9		GGTGTTAG	124826	    7m8
# GGTGTTAG	102580	    7m8		    GAGAGCAA	112390	    AAV9
# ACGAACGA	88528	    AAV2		CGAAGCAA	89246	    AAV8
# CGAAGCAA	66618	    AAV8		ACGAACGA	80863	    AAV2
# GTCCTGAA	484	        AAV6		GTCCTGAA	494	        AAV6

nums1 = {'GTGCATGA': 1029270, 'TGTTGGAG': 365845, 'GACATACG': 293252, 'GAGTAACG': 138779, 'GAGAGCAA': 127729, 'GGTGTTAG': 102580, 'ACGAACGA': 88528, 'CGAAGCAA': 66618, 'GTCCTGAA': 484}
nums2 = {'GTGCATGA': 868583, 'TGTTGGAG': 378761, 'GACATACG': 274134, 'GAGTAACG': 154330, 'GGTGTTAG': 124826, 'GAGAGCAA': 112390, 'CGAAGCAA': 89246, 'ACGAACGA': 80863, 'GTCCTGAA': 494}

# taking average of the two numbers for each barcode
nums = {}
for key in nums1:
    nums[key] = (nums1[key] + nums2[key]) / 2
nums
```

```python
# create a nums array
nums_array = []
# for barcode in data['barcodes'], append the corresponding number from nums into nums_array
for barcode in data['barcodes']:
    nums_array.append(nums[barcode])

# add the nums_array to the data
data['nums'] = nums_array
data.to_excel('fastq_barcode_matrix_raw_counts.xlsx', index=False)
```

```python
# normalize the data in columns of data by the corresponding number in nums
for i in range(2, 11):
    data.iloc[:, i] = data.iloc[:, i] / data['nums']
data.to_excel('fastq_barcode_matrix_normalized_counts.xlsx', index=False)
data
```
