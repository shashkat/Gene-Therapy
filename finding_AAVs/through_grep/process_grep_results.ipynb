### Analysis with larger barcodes

```python
# make a dictionary of barcode names

# Virus     Barcode     Barcode Sequence
# AAV6      Barcode-8   GTCCTGAATCGACTAGAGCT
# AAV8      Barcode-12  CGAAGCAATCGACTAGAGCT
# AAV9      Barcode-13  GAGAGCAATCGACTAGAGCT
# AAV2      Barcode-29  ACGAACGATCGACTAGAGCT
# AAV92YF   Barcode-33  GTGCATGATCGACTAGAGCT
# K912      Barcode-50  ACCACAAGTCGACTAGAGCT
# RH10      Barcode-62  TGTTGGAGTCGACTAGAGCT
# 7m8       Barcode-70  GGTGTTAGTCGACTAGAGCT
# ATX001    Barcode-84  GAGTAACGTCGACTAGAGCT
# ATX002    Barcode-88  GACATACGTCGACTAGAGCT

barcode_names = {'GTCCTGAATCGACTAGAGCT': 'AAV6', 'CGAAGCAATCGACTAGAGCT': 'AAV8', 'GAGAGCAATCGACTAGAGCT': 'AAV9', 'ACGAACGATCGACTAGAGCT': 'AAV2', 'GTGCATGATCGACTAGAGCT': 'AAV92YF', 'ACCACAAGTCGACTAGAGCT': 'K912', 'TGTTGGAGTCGACTAGAGCT': 'RH10', 'GGTGTTAGTCGACTAGAGCT': '7m8', 'GAGTAACGTCGACTAGAGCT': 'ATX001', 'GACATACGTCGACTAGAGCT': 'ATX002'}

# create a barcode_names_array array
barcode_names_array = []
for barcode in data['barcodes']:
    barcode_names_array.append(barcode_names[barcode])
# add the barcode_names_array to the data and put it in the second column
data.insert(1, 'barcode_names', barcode_names_array)
data
```

```python
# remove the zeroth row: with barcode ACCACAAGTCGACTAGAGCT, as it doesnt have normalization values
data = data[data.barcodes != 'ACCACAAGTCGACTAGAGCT']
data
```

```python
# below is the information about the initial numbers for each barcode's virus 
# notice the discrepancy between the order of the barcodes in the two replicates

# GTGCATGATCGACTAGAGCT	1029270		AAV92YF		GTGCATGATCGACTAGAGCT	868583		AAV92YF
# TGTTGGAGTCGACTAGAGCT	365845		Rh10		TGTTGGAGTCGACTAGAGCT	378761		Rh10
# GACATACGTCGACTAGAGCT	293252		ATX002		GACATACGTCGACTAGAGCT	274134		ATX002
# GAGTAACGTCGACTAGAGCT	138779	    ATX001		GAGTAACGTCGACTAGAGCT	154330	    ATX001
# GAGAGCAATCGACTAGAGCT	127729	    AAV9		GGTGTTAGTCGACTAGAGCT	124826	    7m8
# GGTGTTAGTCGACTAGAGCT	102580	    7m8		    GAGAGCAATCGACTAGAGCT	112390	    AAV9
# ACGAACGATCGACTAGAGCT	88528	    AAV2		CGAAGCAATCGACTAGAGCT	89246	    AAV8
# CGAAGCAATCGACTAGAGCT	66618	    AAV8		ACGAACGATCGACTAGAGCT	80863	    AAV2
# GTCCTGAATCGACTAGAGCT	484	        AAV6		GTCCTGAATCGACTAGAGCT	494	        AAV6

nums1 = {'GTGCATGATCGACTAGAGCT': 1029270, 'TGTTGGAGTCGACTAGAGCT': 365845, 'GACATACGTCGACTAGAGCT': 293252, 'GAGTAACGTCGACTAGAGCT': 138779, 'GAGAGCAATCGACTAGAGCT': 127729, 'GGTGTTAGTCGACTAGAGCT': 102580, 'ACGAACGATCGACTAGAGCT': 88528, 'CGAAGCAATCGACTAGAGCT': 66618, 'GTCCTGAATCGACTAGAGCT': 484}
nums2 = {'GTGCATGATCGACTAGAGCT': 868583, 'TGTTGGAGTCGACTAGAGCT': 378761, 'GACATACGTCGACTAGAGCT': 274134, 'GAGTAACGTCGACTAGAGCT': 154330, 'GGTGTTAGTCGACTAGAGCT': 124826, 'GAGAGCAATCGACTAGAGCT': 112390, 'CGAAGCAATCGACTAGAGCT': 89246, 'ACGAACGATCGACTAGAGCT': 80863, 'GTCCTGAATCGACTAGAGCT': 494}

# taking average of the two numbers for each barcode
nums = {}
for key in nums1:
    nums[key] = (nums1[key] + nums2[key]) / 2
nums
```

```python
# create a nums array
nums_array = []
# for barcode in data['barcodes'], append the corresponding number from nums into nums_array
for barcode in data['barcodes']:
    nums_array.append(nums[barcode])
```

```python
# add the nums_array to the data
data['nums'] = nums_array
data.to_excel('fastq_barcode_matrix_raw_counts.xlsx', index=False)
```

```python
# normalize the data in columns of data by the corresponding number in nums
for i in range(2, 11):
    data.iloc[:, i] = data.iloc[:, i] / data['nums']
data.to_excel('fastq_barcode_matrix_normalized_counts.xlsx', index=False)
data
```
